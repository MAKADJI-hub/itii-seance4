```python
import numpy as np
def racine(b,n,e):
    r = e
    while r**n < b:
        r+=e
        print(np.round(r,2))

racine(16,4,0.01)
```

```python
from math import sqrt

def polynome(a, b, c, x):
    print ("En", x, ":", a * x * x + b * x + c)

def racine(a, b, c):
    print("Solutions de ", a, "x^2 +", b, "x +", c, "= 0")
    if a == 0:
        if b == 0:
            if c == 0:
                print ("Tout r ́eel est solution")
            else :
                print ("Aucune solution")
        else :
            x = (-c * 1.0) / b
            print ("Une solution :", xpolynome(a, b, c, x))
    else :
        d = b * b - 4 * a * c
    if(d == 0): 
        x = -b / (2.0 * a) 
        print ("Une solution :", xpolynome(a, b, c, x) )
    elif(d > 0):
        r = sqrt(d) 
        x1 = (-b - r) / (2.0 * a) 
        x2 = (-b + r) / (2.0 * a)
        print ("Deux solutions :", x1, "et", x2)
        polynome(a, b, c, x1) 
        polynome(a, b, c, x2)
    else :
        print ("Aucune solution")

print(racine(3,2,1))
```
